<h6><i>Introduction</i></h6>
<p style="text-align: center;text-shadow: 1px 2px 2px #CE5937;"><b>Breadth First Search</b></p>
<br>
<p><b>Breadth-First-Search(BFS)</b>is an algorithm used for tree traversal on graphs or tree data structure ,BFS can be easily implemented using recursionand data structures like dictionaries and lists 
 <h3 style="text-shadow: 1px 2px 2px #CE5937;"> The Algorithm</h3>
 <ol>
    <li>pick any node .visit the adjacent unvisited vertex mark it as visited,display it and insert it in queue  </li>
    <li> If there are no remaining adjacent vertices left, remove the first vertex from the queue. </li>
    <li>Repeat setp 1 and step 2 untill the queue is empty or the desired node is found </li>
    </ol>
     <h3 style="text-shadow: 1px 2px 2px #CE5937;"> Implementation</h3>
     <p> Consider this graph implemented in the code below:</p>
     

```python
graph = {
  'A' : ['B','C'],
  'B' : ['D', 'E'],
  'C' : ['F'],
  'D' : [],
  'E' : ['F'],
  'F' : []
}

visited = [] # List to keep track of visited nodes.
queue = []     #Initialize a queue

def bfs(visited, graph, node):
  visited.append(node)
  queue.append(node)

  while queue:
    s = queue.pop(0) 
    print (s, end = " ") 

    for neighbour in graph[s]:
      if neighbour not in visited:
        visited.append(neighbour)
        queue.append(neighbour)

# Driver Code
bfs(visited, graph, 'A')
```
